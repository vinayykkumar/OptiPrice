Key Variables Analysis (KVP) - Demand Analysis, Competitor Analysis, Data Type Conversions, and Missing Value Analysis

```python
#KVP1: Demand Analysis

# Group by product and calculate total sales
demand = df.groupby("Product_ID")["Order_ID"].count().reset_index()
demand.rename(columns={"Order_ID": "Total_Orders"}, inplace=True)

# Plot demand distribution
plt.figure(figsize=(8, 5))
sns.histplot(demand["Total_Orders"], bins=30, kde=True, color="blue")
plt.title("Demand Distribution")
plt.xlabel("Total Orders")
plt.ylabel("Frequency")
plt.show()
```

```python
#Demand Analysis (Total Sales by Product(category))

# Calculate total sales per product
total_sales = df.groupby("Category")["Discounted_Price"].sum().reset_index()
total_sales.rename(columns={"Discounted_Price": "Total_Sales"}, inplace=True)

# Display top products by sales
print(total_sales.sort_values(by="Total_Sales", ascending=False).head(10))
```

```python
#Total Sales per Product

total_sales = df.groupby("Product_ID")["Discounted_Price"].sum().reset_index()
total_sales.rename(columns={"Discounted_Price": "Total_Sales"}, inplace=True)

# Print top 15 products by total sales
print("Total Sales per Product:")
print(total_sales.sort_values(by="Total_Sales", ascending=False).head(15))

```

```python
 #Average Sales per Product

average_sales = df.groupby("Product_ID")["Discounted_Price"].mean().reset_index()
average_sales.rename(columns={"Discounted_Price": "Average_Sales"}, inplace=True)

# Print top 15 products by average sales
print("\nAverage Sales per Product:")
print(average_sales.sort_values(by="Average_Sales", ascending=False).head(15))
```

```python
#KVP2: Competitor Analysis
# Compare average prices between platform and competitors
df["Price_Difference"] = df["Original_Price"] - df["Competitor_Price"]

# Boxplot for price differences
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Category"], y=df["Price_Difference"])
plt.title("Price Difference Between Platform and Competitors")
plt.xticks(rotation=45)
plt.show()

# Identify overpriced & underpriced products
overpriced = df[df["Price_Difference"] > 0].shape[0]
underpriced = df[df["Price_Difference"] < 0].shape[0]

print(f"Overpriced Products: {overpriced}, Underpriced Products: {underpriced}")
```

```python
#Competitor Analysis (Average Price Comparison)

# Calculate average prices for platform and competitors
avg_prices = df.groupby("Category").agg(
    Platform_Avg_Price=("Discounted_Price", "mean"),
    Competitor_Avg_Price=("Competitor_Price", "mean")
).reset_index()

# Calculate price difference
avg_prices["Price_Difference"] = avg_prices["Platform_Avg_Price"] - avg_prices["Competitor_Avg_Price"]

# Display results
print(avg_prices)
```

```python
#KVP3: Data Type Conversions
# Convert data types
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Delivery_Time (Days)"] = pd.to_numeric(df["Delivery_Time (Days)"], errors="coerce")

# Convert categorical variables to category type
df["Category"] = df["Category"].astype("category")
df["Brand"] = df["Brand"].astype("category")

# Check updated data types
print(df.dtypes)
```

```python
#KVP4: Missing Value Analysis & Treatment

# Check missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values[missing_values > 0])

# Fill missing values
df["Ratings"] = df["Ratings"].fillna(df["Ratings"].median())  # Fill with median
df["Discount (%)"] = df["Discount (%)"].fillna(0)  # Assume no discount if missing
df = df.fillna(method="ffill")  # Forward fill for categorical data

# Verify missing values removed
print("Remaining Missing Values:\n", df.isnull().sum().sum())
```
